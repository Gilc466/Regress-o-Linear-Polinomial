# **Regressão Polinomial**

---



Há situações em que a relação entre as variáveis independentes e a variável dependente é não-linear, fazendo com que os modelos de regressão linear simples ou múltipla tenham dificuldades com esse cenário.

Outrossim, destaca-se que, teoricamente, podemos aproximar qualquer função com um polinômio. Sendo assim, existe a alternativa do modelo de **Regressão Polinomial**.

A partir das variáveis existentes, novas variáveis polinomiais vão sendo geradas e a regressão com elas terá mais capacidade quanto maior o grau do polinômio criado. Dessa forma, o algoritmo de regressão não se modifica, pois estamos apenas criando novas variáveis polinomiais.

Equação:

\begin{equation}
   y = b_0 + b_1 ⋅ x_1 +  b_2 ⋅ x_1^2 + \cdots + b_n ⋅ x_1^n
\end{equation}
sendo que:
*   $y$ é a variável dependente;
*   $x_1$ é a variável dependente;
*   $b_0, b_1, ⋯, b_n$ são os coeficientes do modelo.

Umas das limitações dos modelos de regressão é que a máquina não aprende a forma da não linearidade, sendo que é preciso forjar variáveis novas para essa classe de modelo linear. 

Outro aspecto sobre o modelo de regressão polinomial é que ele não funciona bem no caso de muitas variáveis e com elevado grau do polinômio. A capacidade de ajuste aos dados é maior conforme aumentamos o grau do polinômio, porém, provavelmente vai prejudicar a performance de generalização, apresentando o problema de sobreajuste.

Portanto, é recomendável utilizar a regressão polinomial em casos de não linearidades mais simples, considerando um polinômio de grau dois ou no máximo 3.

# Exemplo

## 1. Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 2. Importando os conjuntos de dados

```python
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

## 3. Treinando o Modelo de Regressão Linear

```python
from sklearn.linear_model import LinearRegression
reg_simples = LinearRegression()
reg_simples.fit(X, y)
```

## 4. Treinando o Modelo de Regressão Polinomial

```python
from sklearn.preprocessing import PolynomialFeatures
form_variaveis = PolynomialFeatures(degree = 2)
X_poly = form_variaveis.fit_transform(X)
reg_polinomial = LinearRegression()
reg_polinomial.fit(X_poly, y)

np.set_printoptions(suppress=True) # suprimir a notação científica
```

## 5. Visualizando os Resultados da Regressão Linear Simples

```python
plt.figure(figsize=(8, 5))

plt.scatter(X, y, color='red', s=60, edgecolor='black', label='Dados Reais')  # Aumentar tamanho e adicionar borda aos pontos

# Linha da regressão
plt.plot(X, reg_simples.predict(X), color='blue', linewidth=2, label='Linha de Regressão')

plt.title('Regressão Linear', fontsize=15)
plt.xlabel('X', fontsize=12)
plt.ylabel('y', fontsize=12)

plt.grid(True, which='both', linestyle='--', linewidth=0.7, color='gray')
plt.legend(fontsize=10)

plt.show()
```

## 6. Visualizando os Resultados da Regressão Polinomial

```python
plt.figure(figsize=(8, 5))  # Ajustar o tamanho da figura

# Gráfico de dispersão dos dados reais
plt.scatter(X, y, color='red', s=60, edgecolor='black', label='Dados Reais')  

# Gráfico de dispersão dos dados previstos pela regressão polinomial
plt.scatter(X, reg_polinomial.predict(form_variaveis.fit_transform(X)), color='blue', s=40, label='Previsões Polinomiais')

plt.title('Regressão Polinomial', fontsize=15)
plt.xlabel('X', fontsize=12)
plt.ylabel('y', fontsize=12)

plt.grid(True, which='both', linestyle='--', linewidth=0.7, color='gray')

plt.legend(fontsize=10)
plt.show()
```

## 7. Previsão de resultados para o modelo

```python
reg_simples.predict([[6.5]])  #regressão linear simples
```

```python
reg_polinomial.predict(form_variaveis.fit_transform([[6.5]])) #regressão polinomial
```
